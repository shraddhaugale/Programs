```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([23,34,42,49,55,61,68,73,78,82])
y=np.array([25,29,33,37,44,47,57,62,67,71])
z=np.array([87,82,78,73,67,63,55,51,47,39])
```

```python
## Line Plot
```

```python
plt.title('line plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='yellow',color='black',ls='--',ms=5,label='2023'); #ls= line style
plt.plot(x,z,marker='H',mfc='blue',color='black',ls='--',ms=5,label='2024');
plt.grid()
plt.legend();
```
